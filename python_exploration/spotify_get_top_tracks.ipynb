```python
import spotipy
import spotipy.util as util
from config import client_id, client_secret
import pandas as pd
```

```python
redirect_uri = 'http://localhost:7777/callback'
scope = 'user-top-read'
```

```python
token = util.prompt_for_user_token(
                                   scope=scope, 
                                   client_id=client_id,   
                                   client_secret=client_secret,     
                                   redirect_uri=redirect_uri)
```

```python
sp = spotipy.Spotify(auth=token)
```

```python
top_tracks = sp.current_user_top_tracks(limit=15,time_range='long_term')
```

```python
top_tracks
```

```python
#getting the id of a song
top_tracks["items"][0]["id"]
```

```python
#putting the name of a song in the list
top_songs =[]
for idx, item in enumerate(top_tracks['items']):
    top_songs.append(item['id'])
```

```python
#viewing list of top 12 songs 
top_songs
```

```python
id_string = ','.join(top_songs)
```

```python
id_string
```

```python
test_features = sp.audio_features(id_string)
```

```python
#getting audio features of those top 5 songs
features = []
for id in top_songs:
    features.append(sp.audio_features(id))
```

```python
#viewing features
features
```

```python
#count of top songs
song_count = len(top_songs)
song_count
```

```python
#setting empty list for all the song features
dance = []
energy = []
key = []
loud = []
mode = []
speech = []
acoustic = []
instrument = []
liveness = []
valence = []
tempo = []
```

```python
#adding all the features from the JSON to the list
for i in range(song_count):
    dance.append(features[i][0]["danceability"])
    energy.append(features[i][0]["energy"])
    key.append(features[i][0]["key"])
    loud.append(features[i][0]["loudness"])
    mode.append(features[i][0]["mode"])
    speech.append(features[i][0]["speechiness"])
    acoustic.append(features[i][0]["acousticness"])
    instrument.append(features[i][0]["instrumentalness"])
    liveness.append(features[i][0]["liveness"])
    valence.append(features[i][0]["valence"])
    tempo.append(features[i][0]["tempo"])
```

```python
#create dictionary from list
song_dict = dict(danceability=dance, 
                 energy =energy,
                 key =key,
                 loudness=loud,
                 mode=mode,
                 speechiness=speech,
                 acousticness=acoustic,
                 instrumentalness=instrument,
                 liveness=liveness,
                 valence=valence,
                 tempo=tempo)
```

```python
#putting dictionary in a dataframe
song_df = pd.DataFrame(song_dict)
song_df
```

```python
top_songs_means = song_df[["danceability","energy","key","loudness",
      "mode","speechiness","acousticness",
      "instrumentalness","liveness","valence","tempo"]].mean()
```

```python
top_mean = top_songs_means.mean()
```
